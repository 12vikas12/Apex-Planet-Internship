```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.graphics.mosaicplot import mosaic
from joypy import joyplot

# Load dataset
df = pd.read_excel("/content/Shopping_data.csv.xlsx")

print("Columns in dataset:")
print(df.columns)

# Stacked Bar Chart
plt.figure(figsize=(12, 6))
df_grouped = df.groupby(['Genre', 'Age']).size().unstack()
df_grouped.plot(kind='bar', stacked=True, colormap='viridis', figsize=(12, 6))
plt.title('Stacked Bar Chart: Age vs Genre', fontsize=16)
plt.ylabel('Count', fontsize=12)
plt.xlabel('Genre', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Grouped Bar Chart
plt.figure(figsize=(12, 6))
sns.countplot(x='Genre', hue='Age', data=df, palette='viridis')
plt.title('Grouped Bar Chart: Age vs Genre', fontsize=16)
plt.ylabel('Count', fontsize=12)
plt.xlabel('Genre', fontsize=12)
plt.xticks(rotation=45)
plt.legend(title='Age')
plt.tight_layout()
plt.show()

# Mosaic Plot
plt.figure(figsize=(12, 6))
mosaic(df, ['Genre', 'Age'], title='Mosaic Plot: Genre and Age')
plt.show()

# Bar Chart (using Spending Score)
plt.figure(figsize=(12, 6))
score_summary = df.groupby('Genre')['Spending Score (1-100)'].agg(['mean', 'median', 'std']).reset_index()
sns.barplot(x='Genre', y='mean', data=score_summary, palette='coolwarm', ci=None)
plt.title('Bar Chart: Genre vs Spending Score Mean', fontsize=16)
plt.ylabel('Average Spending Score', fontsize=12)
plt.xlabel('Genre', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# KDE Plot
plt.figure(figsize=(12, 6))
sns.kdeplot(data=df, x='Spending Score (1-100)', hue='Genre', multiple='stack', palette='coolwarm')
plt.title('Grouped KDE Plot: Genre vs Spending Score', fontsize=16)
plt.xlabel('Spending Score', fontsize=12)
plt.ylabel('Density', fontsize=12)
plt.show()

# Box Plot
plt.figure(figsize=(12, 6))
sns.boxplot(x='Genre', y='Age', data=df, palette='Set2')
plt.title('Box Plot: Genre vs Age', fontsize=16)
plt.ylabel('Age', fontsize=12)
plt.xlabel('Genre', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Violin Plot
plt.figure(figsize=(12, 6))
sns.violinplot(x='Genre', y='Spending Score (1-100)', data=df, palette='Set3')
plt.title('Violin Plot: Genre vs Spending Score', fontsize=16)
plt.ylabel('Spending Score', fontsize=12)
plt.xlabel('Genre', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Ridgeline Plot
plt.figure(figsize=(12, 8))
joyplot(data=df, by='Genre', column='Spending Score (1-100)', colormap=plt.cm.coolwarm, fade=True)
plt.title('Ridgeline Plot: Genre vs Spending Score', fontsize=16)
plt.xlabel('Spending Score', fontsize=12)
plt.show()

# Beeswarm Plot
plt.figure(figsize=(12, 6))
sns.swarmplot(x='Genre', y='Spending Score (1-100)', data=df, palette='husl', size=8)
plt.title('Beeswarm Plot: Genre vs Spending Score', fontsize=16)
plt.ylabel('Spending Score', fontsize=12)
plt.xlabel('Genre', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

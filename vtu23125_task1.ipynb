```python
import matplotlib.pyplot as plt
import pandas as pd
data = pd.read_excel("/content/Iris (1).csv.xlsx")
print(data.columns)
plt.scatter(data['PetalLengthCm'], data['SepalLengthCm'])
plt.xlabel('Petal Length (cm)')
plt.ylabel('Sepal Length (cm)')
plt.title('Petal Length vs Sepal Length')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
data = pd.read_excel("/content/Iris (1).csv.xlsx")
print(data.columns)
sns.barplot(x=data['PetalLengthCm'], y=data['PetalWidthCm'])
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.title('Petal Length vs Petal Width')
plt.show()
```

```python
import plotly.express as px
import pandas as pd

data = pd.read_excel("/content/Iris (1).csv.xlsx")
df = pd.DataFrame(data)

species_counts = df[df.columns[-1]].value_counts().reset_index()
species_counts.columns = ['species', 'count']

fig = px.bar(species_counts, x='species', y='count')
fig.show()
```

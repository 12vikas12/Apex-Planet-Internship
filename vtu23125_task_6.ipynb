```python
import pandas as pd
import matplotlib.pyplot as plt
import networkx as nx
import kagglehub


# Using your dataset (df already contains: 'Category', 'Item Purchased', 'Payment Method', 'Shipping Type')
path = kagglehub.dataset_download("iamsouravbanerjee/customer-shopping-trends-dataset")
df = pd.read_csv(path + "/shopping_trends.csv")

G_cat_item = nx.DiGraph()
for _, row in df.iterrows():
    G_cat_item.add_edge(row['Category'], row['Item Purchased'])

pos1 = nx.spring_layout(G_cat_item, k=0.7, iterations=70, seed=42)
plt.figure(figsize=(18, 12))
nx.draw(G_cat_item, pos1, with_labels=True, node_color='lightcoral', node_size=1200,
        font_size=11, font_weight='bold', edge_color='gray', arrows=True)
plt.title('Force-Based Layout: Category → Item Purchased', fontsize=18, fontweight='bold')
plt.show()

G_pay_ship = nx.DiGraph()
for _, row in df.iterrows():
    G_pay_ship.add_edge(row['Payment Method'], row['Shipping Type'])

pos2 = nx.spring_layout(G_pay_ship, k=0.7, iterations=70, seed=42)
plt.figure(figsize=(18, 12))
nx.draw(G_pay_ship, pos2, with_labels=True, node_color='skyblue', node_size=1200,
        font_size=11, font_weight='bold', edge_color='gray', arrows=True)
plt.title('Force-Based Layout: Payment Method → Shipping Type', fontsize=18, fontweight='bold')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import networkx as nx
import kagglehub

# Download and load the Iris dataset
path = kagglehub.dataset_download("kamrankausar/iris-data")
print("Path to dataset files:", path)

# Load the dataset (check for the correct filename)
df = pd.read_csv(path + "/Iris.csv")

# Display basic info
print("Columns available:", df.columns.tolist())
print(df.head())

# Force-based layout 1: Relationship between Species → SepalWidthCm
G_species_sepal = nx.DiGraph()
for _, row in df.iterrows():
    G_species_sepal.add_edge(row['Species'], str(row['SepalWidthCm']))

pos1 = nx.spring_layout(G_species_sepal, k=0.7, iterations=70, seed=42)
plt.figure(figsize=(14, 10))
nx.draw(G_species_sepal, pos1, with_labels=True, node_color='lightgreen', node_size=1500,
        font_size=12, font_weight='bold', edge_color='gray', arrows=True)
plt.title('Force-Based Layout: Species → Sepal Width (cm)', fontsize=18, fontweight='bold')
plt.show()

# Force-based layout 2: Relationship between Species → PetalLengthCm
G_species_petal = nx.DiGraph()
for _, row in df.iterrows():
    G_species_petal.add_edge(row['Species'], str(row['PetalLengthCm']))

pos2 = nx.spring_layout(G_species_petal, k=0.7, iterations=70, seed=42)
plt.figure(figsize=(14, 10))
nx.draw(G_species_petal, pos2, with_labels=True, node_color='skyblue', node_size=1500,
        font_size=12, font_weight='bold', edge_color='gray', arrows=True)
plt.title('Force-Based Layout: Species → Petal Length (cm)', fontsize=18, fontweight='bold')
plt.show()
```
